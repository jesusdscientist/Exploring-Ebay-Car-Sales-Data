## Exploring Price by Brand

```python
autos["brand"].unique()
```

```python
brand_counts = autos["brand"].value_counts(normalize = True)
a = brand_counts[brand_counts > 0.05].index
```

```python
aggragated_data = {}
for s in a:
    selected_row = autos[autos["brand"] == s]
    mean_brand = selected_row["price_$"].mean()
    aggragated_data[s] = int(mean_brand)
```

```python
aggragated_data
```

## Exploring Mileage

```python
bmp_series = pd.Series(aggragated_data)
```

```python
df = pd.DataFrame(bmp_series, columns = ["mean_price"])
df
```
